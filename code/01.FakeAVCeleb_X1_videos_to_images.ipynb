### 1. save `REAL_A` to `real` folder

- total **REAL** images: 390 videos * 25 frames = 9750

```python
folders = glob.glob('/content/drive/MyDrive/American_University/2021_Fall/DATA-642-001_Advanced Machine Learning/data/FakeAVCeleb/FakeAVCeleb_preprocessed/TRAIN/EXTRACTED/VIDEOS/REAL_A/*')
list_real_filenames = []
count
for folder in folders:
    for f in glob.glob(folder+'/*.mp4'):
        list_real_filenames.append(f)

for i in range(len(list_real_filenames)):
  cam = cv2.VideoCapture(list_real_filenames[i])
  try:   
      # creating a folder named images
      if not os.path.exists('data'):
          os.makedirs('data')
  # if not created then raise error
  except OSError:
      print ('Error: Creating directory of images data')

  count = 0  # count the number of pictures
  frame_interval = 1 # video frame count interval frequency
  frame_interval_count = 0
  currentframe = 1 # start with frame 1

  while(True):
    # reading from frame
    ret,frame = cam.read()
    if ret:
      if frame_interval_count % frame_interval == 0 and currentframe <= 25:
      # if video is still left continue creating images, extract up to 10 imgs /per video
        name = 'REAL_A' + str(i) + "_" + str(currentframe) + '.jpg'
        print ('Creating...' + name)
        frame_interval_count += 1
      # writing the extracted images
        cv2.imwrite("/content/drive/MyDrive/American_University/2021_Fall/DATA-642-001_Advanced Machine Learning/data/data_ready/images/real/"+name, frame)
        count += 1
      # increasing counter so that it will show how many frames are created
        currentframe += 1
      else:
        frame_interval_count += 1
    else:
      break
#print(count)
# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```
