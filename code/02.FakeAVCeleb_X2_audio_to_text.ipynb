### 1. save `REAL_A` to `real` folder

- total **REAL** text: 390 videos to 390 text files

```python
# should be mounting the Google Drive first
folders = glob.glob('/content/drive/MyDrive/American_University/2021_Fall/DATA-642-001_Advanced Machine Learning/data/FakeAVCeleb/FakeAVCeleb_preprocessed/TRAIN/EXTRACTED/AUDIOS/REAL_A/*')
list_real_filenames = []
for folder in folders:
    for f in glob.glob(folder+'/*.wav'):
        list_real_filenames.append(f)

for i in range(len(list_real_filenames)):
  # initialize the recognizer
  r = sr.Recognizer()
  filename = list_real_filenames[i]
  with sr.AudioFile(filename) as source:
    # listen for the data (load audio to memory)
    audio = r.record(source)
  try:
    text = r.recognize_google(audio)
    print("Text: "+text)
  except Exception as e:
    print("Exception: "+str(e))
    print('Sorry.. texts cannot be captured..')
  %cd /content/drive/MyDrive/American_University/2021_Fall/DATA-642-001_Advanced Machine Learning/data/data_ready/text/real
  with open("REAL_A" + str(i) + '.txt', "w") as f:
    f.writelines(text)
    print("Saving.." + "REAL_A_" + str(i) + '.txt')
```
