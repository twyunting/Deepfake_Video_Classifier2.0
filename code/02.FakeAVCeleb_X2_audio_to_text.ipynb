# Start to translate audios to texts as $X_2$ features
## 1. save `REAL_A` to `real` text folder

- total **REAL** text: 390 audios to 390 text files
- If the `recognizer` cannot captures the text from a given video, the text will denote as "NA"

```python
# should be mounting the Google Drive first
folders = glob.glob('/content/drive/MyDrive/American_University/2021_Fall/DATA-642-001_Advanced Machine Learning/data/data_ready/audios/real')
list_real_filenames = []
for folder in folders:
    for f in glob.glob(folder+'/*.wav'):
        list_real_filenames.append(f)
print(len(list_real_filenames))

for i in range(len(list_real_filenames)):
  # initialize the recognizer
  r = sr.Recognizer()
  filename = list_real_filenames[i]
  with sr.AudioFile(filename) as source:
    # listen for the data (load audio to memory)
    audio = r.record(source)
  try:
    text = r.recognize_google(audio)
    print("Text: "+text)
  except Exception as e:
    print("Exception: "+str(e))
    print('Sorry.. texts cannot be captured..')
  %cd /content/drive/MyDrive/American_University/2021_Fall/DATA-642-001_Advanced Machine Learning/data/data_ready/texts/real
  with open("REAL_A" + str(i) + '.txt', "w") as f:
    f.writelines(text)
    print("Saving.." + "REAL_A" + str(i) + '.txt')
```

## 2. save `FAKE_BCD` to `fake` text folder
- total **FAKE** text: 390 audios to 390 text files
- If the `recognizer` cannot captures the text from a given video, the text will denote as "NA"
- `FAKE_BCD` includes `FAKE_RTVC_B`, `FAKE_FSGAN_C` and `FAKE_FACESWAP_D_W2L`.

```python
# should be mounting the Google Drive first
folders = glob.glob('/content/drive/MyDrive/American_University/2021_Fall/DATA-642-001_Advanced Machine Learning/data/data_ready/audios/fake')
list_fake_filenames = []
for folder in folders:
    for f in glob.glob(folder+'/*.wav'):
        list_fake_filenames.append(f)

for i in range(len(list_fake_filenames)):
  # initialize the recognizer
  r = sr.Recognizer()
  filename = list_fake_filenames[i]
  with sr.AudioFile(filename) as source:
    # listen for the data (load audio to memory)
    audio = r.record(source)
  try:
    text = r.recognize_google(audio)
    print("Text: "+text)
  except Exception as e:
    print("Exception: "+str(e))
    print('Sorry.. texts cannot be captured..')
  %cd /content/drive/MyDrive/American_University/2021_Fall/DATA-642-001_Advanced Machine Learning/data/data_ready/texts/fake
  with open("FAKE_BCD" + str(i) + '.txt', "w") as f:
    f.writelines(text)
    print("Saving.." + "FAKE_BCD" + str(i) + '.txt')
```
